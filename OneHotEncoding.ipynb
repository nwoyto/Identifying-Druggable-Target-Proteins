```python
df.rename( columns={'Unnamed: 0':'DTP', 'Hydrophobicity      ': 'Hydrophobicity'}, inplace=True )
```

```python
dtp_col = df[['DTP','Protein']]
used_data = df.drop(['DTP','Xaa','Glx', 'Asx'], axis = 1)
```

```python
used_data.head()
```

```python
print(used_data.dtypes)
```

```python
# create an empty list to store column names
int_cols = []

# iterate through the columns in the dataframe and append columns of dtype int to int_col
for col in used_data.columns:
    if used_data[col].dtype == 'int64':
        int_cols.append(col)

print(int_cols)
```

```python
cat_df = used_data[int_cols]
cat_df.head(2)
```

```python
def one_hot_encode_cols(df, cols):
    for col in cols:
        df = pd.concat([df, pd.get_dummies(df[col], prefix=col)], axis=1)
        df = df.drop(col, axis=1)
    return df

cat_df = one_hot_encode_cols(used_data, int_cols)
```

```python
cat_df.head()
```

```python
cat_df = cat_df.drop('Protein', axis=1)
cat_df.columns.unique()
```

```python
unique_columns = []

for i in cat_df.columns.unique():
    unique_columns.append(i)

print(unique_columns)
print(len(unique_columns))
```

```python
ohe_df = cat_df
ohe_df.head(2)
```

```python
ohe_df.head(2)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
norm_data = scaler.fit_transform(ohe_df)
norm_data = pd.DataFrame(norm_data, columns=ohe_df.columns)
norm_data.head()
```

```python
norm_data = pd.concat([dtp_col,norm_data],axis=1)
norm_data.head(2)
```
